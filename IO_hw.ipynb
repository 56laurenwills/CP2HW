```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Image
```

```python
integer = (1,2,3,4,5)
#For v_px = v_=py = v_pz = v_pt = float, v_int = integer
compression_1_tree = (1.46, 1.46, 1.46, 1.46, 1.45)
compression_1_variable = (2.79, 2.85, 2.82, 2.85, 2.67)
disk_1 = (0.17381, 0.119935, 0.0908638, 0.092615, 0.124728)
cpu_1 = (37.51, 34.84, 34.15, 34.18, 38.88)
#for v_px = v_py = v_pz = v_pt = v_int = float
compression_2_tree = (1.30, 1.30, 1.30, 1.30, 1.31, 1.29)
compression_2_variable = (1.18, 1.18, 1.18, 1.18, 1.17)
disk_2 = (0.13194, 0.0859, 0.075262, 0.0718, 0.089928)  
cpu_2 = (37.98, 36.46, 35.95, 35.97, 38.15)
```

```python
plt.scatter(integer, disk_1, label = 'four floats, 1 int')
plt.scatter(integer, disk_2, marker = '*', label = 'all floats')
plt.title('Disk usage for all floats versus 4 floats 1 int')
plt.xlabel('Flush size- 0, 30MB, 10MB, 10,000, and 100')
plt.ylabel('Disk usage')
plt.legend()
plt.show()
```

```python
plt.scatter(integer, compression_1_variable, label = 'int')
plt.scatter(integer, compression_2_variable, marker = '*', label = 'floats')
plt.xlabel('Flush size- 0, 30MB, 10MB, 10,000, and 100')
plt.ylabel('Compression factor')
plt.title('The compression factor of the vector changed form int to float')
plt.legend()
plt.show()
```

Results- As seen in the graphs- larger basket sizes means writing out to the disk more, so they have a smaller disk usages, which is accuratley shown (except the 30MB is larger than 10MB and I cannot figure that out, but the 0, 100, and 10,000 basket size agrees with what you would expect). Changing the v_int vector from a integer to a float makes the compression factor of that vector worse, and lowers the compression factor of the entire tree. Also disk usage is lower for data containing only one type of data, even though the compression is worse.
